```python
from LightPipes import *
import matplotlib.pyplot as plt
import imageio
import numpy as np
from PIL import Image
```

```python
image=imageio.imread('4.png')[359:471,359:471]
image1=Image.fromarray(image)
image1=image1.resize((28,28),resample=0)
image2=np.array(image1)
plt.imshow(image2)
plt.colorbar()
plt.show()
```

```python
Field=Begin(38*mm, 0.65*um, 500)
for i in range(28):
    for j in range(28):
        if image2[j,i]<100:
            Field = RectScreen(76*um,76*um, (i-14)*76*um, (j-14)*76*um,0, Field)
Field = RectAperture(2*mm,2*mm, 0*um, 0*um,0, Field)
I=Intensity(0,Field)
plt.imshow(I)
plt.colorbar()
plt.show()
```

```python
Field1 = Forvard(0.164*m, Field)
I = Intensity(0,Field1)
plt.imshow(I)
plt.colorbar()
plt.show()
image=np.array(I)
plt.imshow(image[200:300,200:300])
plt.colorbar()
plt.show()
```

```python
Field2=Lens(0.164*m,0,0,Field1)
Field2 = Forvard(0.15*m, Field2)
I = Intensity(0,Field2)
plt.imshow(I)
plt.colorbar()
plt.show()
image=np.array(I)
plt.imshow(image[200:300,200:300])
plt.colorbar()
plt.show()
```

```python
Field2 = Forvard(0.2*m, Field2)
I = Intensity(0,Field2)
plt.imshow(I)
plt.colorbar()
plt.show()
Field2=Lens(0.2*m,0,0,Field2)
Field2 = Forvard(0.2*m, Field2)
I = Intensity(0,Field2)
plt.imshow(I)
plt.colorbar()
plt.show()
image=np.array(I)
plt.imshow(image[200:300,200:300])
plt.colorbar()
plt.show()
```
