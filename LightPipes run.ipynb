```python
import matplotlib.pyplot as plt
import imageio
import numpy as np
from PIL import Image
from LightPipes import *
```

```python
image=imageio.imread('4.png')[359:471,359:471]
image1=Image.fromarray(image)
image1=image1.resize((50,50),resample=0)
image3=np.array(image1)

for i in range(50):
    for j in range(50):
        if image3[i,j]>50:
            image3[i,j]=1
        else:
            image3[i,j]=0
            
plt.imshow(image3)
plt.colorbar()
plt.show()
```

```python
Field=Begin(173.25*um, 0.63*um, 550)
for i in range(50):
    for j in range(50):
        if image3[j,i]<1:
            Field = RectScreen(0.315*um,0.315*um, (i-25)*0.315*um, (j-25)*0.315*um,0, Field)
Field = RectAperture(15*um,15*um, 0*um, 0*um,0, Field)
I=Intensity(0,Field)
plt.imshow(I)
plt.colorbar()
plt.show()
```
